```python
#Задание 5 (необязательное)
#Дана переменная, в которой хранится шестизначное число (номер проездного билета). 
#Напишите программу, которая будет определять, является ли данный билет "счастливым". 
#Билет считается счастливым, если сумма первых трех цифр совпадает с суммой последних трех цифр номера.
a = int(input("Введите номер билета (6 цифр) "))
a1 = a // 1000
a2 = a % 1000
sum1 = 0
sum2 = 0
while(a1 != 0): #До такой схемы догадаться сама бы не смогла, смотрела пример в интернете.
    sum1 = sum1 + a1 % 10
    a1 = a1 // 10
while(a2 != 0):
    sum2 = sum2 + a2 % 10
    a2 = a2 // 10
#print(sum1, sum2)
if sum1 == sum2 :
    print("Ваш билет счастливый!")
else:
    print("Повезёт в другой раз!")
```

```python
#Задание 6 (необязательное)
#Напишите программу, которая сможет вычислять площади трех фигур (круг, треугольник и прямоугольник). 
#Тип фигуры запрашиваем через пользовательский ввод, после чего делаем запрос характеристик фигуры:
#если пользователь выбрал круг, запрашиваем его радиус,
#если треугольник – длины трех его сторон;
#если прямоугольник – длины двух его сторон.
f = input("Введите тип фигуры (круг, треугольник, прямоугольник) ")
if f == 'Круг' or f == 'круг' or f == 'КРУГ':
    r = int(input("Введите радиус, в см "))
    s1 = 3.1416 * r**2
    print("Площадь круга - ",s1)
elif f == 'Треугольник' or f == 'треугольник' or f == 'ТРЕУГОЛЬНИК':
    a = int(input("Введите длину стороны a, в см "))
    b = int(input("Введите длину стороны b, в см "))
    c = int(input("Введите длину стороны c, в см "))
    s2 = 0.25 * ((a+b+c)*(b+c-a)*(a+c-b)*(a+b-c))**(1/2)
    print("Площадь треугольника - ",s2)
elif f == 'Прямоугольник' or f == 'прямоугольник' or f == 'ПРЯМОУГОЛЬНИК': 
    a2 = int(input("Введите длину стороны a, в см "))
    b2 = int(input("Введите длину стороны b, в см "))
    s3 = a2*b2
    print("Площадь прямоугольника - ",s3)
else:
    print("Неправильно задан тип фигуры")
```
